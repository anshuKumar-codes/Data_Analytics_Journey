```python
df5={
    "Name": ["Anshu",None,"Badal", "Aadil","Deepak","Akash","Anil","Aditya","Ayush"],
    "Age":[28,None,30,20,25,34,21,55,34],
    "Salary":[10000,20000,30000,40000,50000,60000,70000,80000,90000],
    "Performance":[32,44,55,66,78,21,69,88,90],
    " 
}
df5=pd.DataFrame(df5)
print("Sample Dataframe")
df5
```

```python
#---------------------1.Detect NaN value-------------------------
# isnull is use for find NaN value in dataset
df5.isnull().sum()
```

```python
# How to drop NaN values in columns and row 
# When we can give axis in this method then by default is remove all the NaN values in dataset.
df5.dropna(inplace=True)
df5
```

```python
df5={
    "Name": ["Anshu",None,"Badal", "Aadil","Deepak","Akash","Anil","Aditya","Ayush"],
    "Age":[28,None,30,20,25,34,21,55,34],
    "Salary":[10000,20000,None,40000,50000,60000,70000,80000,90000],
    "Performance":[32,44,55,66,78,21,69,88,90],
    
}
df5=pd.DataFrame(df5)
print("Sample Dataframe")
df5
```

```python
df5["Salary"].fillna(df5["Salary"].mean(),inplace=True)
df5
```

```python
#--------------------------- Interpolition --------------------------------
# Preserved data integrity 
# Smooth trends
# avoid loss
# -------------------------------------------------------------------------
# 1-timer series data
# 2-Numeric data with trends
# 3-Avoid dropping rows

df5={
    "Name": ["Anshu",None,"Badal", "Aadil","Deepak","Akash","Anil","Aditya","Ayush"],
    "Age":[28,None,30,20,25,34,21,55,34],
    "Salary":[10000,20000,None,40000,50000,60000,70000,80000,90000],
    "Performance":[32,44,55,66,78,21,69,88,90],
    
}
df5=pd.DataFrame(df5)
print("Sample Dataframe")
df5
```

```python
df5["Salary"].fillna(df5["Salary"].interpolate(),inplace=True)
```
