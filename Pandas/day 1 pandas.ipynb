```python
#--------------------------------- bar graph ------------------------------

# plt.bar(x-axis,y-axis,height,color="color name",width=value,label="labelname")

products=["A","B","C","D"]
sales=[1000,1500,800,1200]
#--------------------------------------------------------------------------------------

colors=[]
for s in sales:
    if s==800:
        colors.append("red")
    elif s>=800 and s<=1200:
        colors.append("yellow")
    elif s==1500:
        colors.append("green")
    
#--------------------------------------------------------------------------------------       

plt.figure(figsize=(12,7))
plt.bar(products,sales,color=colors,label="Sales 2025")
plt.xlabel("Products",size=20,color="red")
plt.ylabel("Sales",size=20,color="blue")
plt.title("Products sales data",size=30,color="green")
plt.grid(color="gray",linestyle=":",linewidth=1)
plt.legend(loc="upper right",labelcolor="Orange")
plt.show()
```

```python
#--------------------------------- pie chart --------------------------------

import matplotlib.pyplot as plt

#plt.pie(x,color=["colors name"],labels=("acc. to your data column name"),autopct="%1.1f%%")
#autopct=("%1.1f%%) 
#"%1.1f means one number before decimal and one number after decimal
# %% means that it show tha sign of percentage in piechart according to numbers

regions=["North","South","East","West"]
revenue=[1000,1500,800,1200]

plt.pie(revenue,labels=regions,colors=["lightgreen","gold","coral","skyblue"],autopct="%1.1f%%")
plt.title("Revenue contribution by region")
plt.show()
```
