```python
import pandas as pd
ipl={'Team':['Riders','Riders','Devils','Devils','Kings','Kings','Kings','Kings','Riders','Royals','Royals','Riders'],
     'Rank':[1,2,2,3,3,4,1,1,2,4,1,2],
     'Year':[2014,2015,2014,2015,2014,2015,2016,2017,2016,2014,2015,2017],
     'Points':[876,789,863,673,741,812,756,788,649,701,804,690]}
df=pd.DataFrame(ipl)
df
```

```python
h=df.groupby('Team')
h
```

```python
p=h['Points'].sum()
p
```

```python
g=pd.DataFrame(p)
g
```

```python
n=g.reset_index()
n
```

```python
t=n['Team']
t2=list(t)
t2
```

```python
s=n['Points']
s2=list(s)
s2
```

```python
k=['blue','red','green','yellow']
```

```python
plt.figure(figsize=(15,3))
plt.barh(t2,s2,color=k)
plt.xlabel('Team',size=20,color='red',fontweight='bold')
plt.ylabel('Points',size=20,color='red',fontweight='bold')
plt.xticks(rotation=30,size=12,color='green')
plt.yticks(rotation=30,size=12,color='green')
plt.title('Team Records Report',size=20,color='red',fontweight='bold')

for index,value in enumerate(s2):
    plt.text(value,index,str(value))

plt.show()
```
